```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

Result = pd.read_csv("CorrEffect_algo.csv", index_col=False)
Result = Result.drop(Result.columns[0], axis=1)

sns.set_theme()
sns.set_style("ticks")
sns.set_context("paper")

Result = Result.set_index(['Correlation'])
print(Result.head())

markers = ['^', 'o', 's']
colors = sns.color_palette('tab10', n_colors=3)
ax,fig = plt.subplots(figsize = (5,3))
ax = sns.scatterplot(data=Result, markers=markers, palette=colors, edgecolor='black', alpha=0.8, s=55)
#ax.set(xlabel=r'$\rho$', ylabel='MSE')
ax.tick_params(axis='both', which='major', labelsize=13)
ax.set_xlabel(r'$\rho$', fontsize=14)
ax.set_ylabel('MSE', fontsize=14)
plt.xticks([-0.6, -0.5, -0.25, 0, 0.25, 0.5, 0.75, 1], fontsize=10.5)
plt.yticks(fontsize=10.5)
plt.xlim(-0.55, 1.05)
plt.grid()

handles, labels = ax.get_legend_handles_labels()
legend = ax.legend(handles, ['SP', 'RA', 'DR'], loc='upper right',edgecolor='black', fontsize=12)
for i in range(len(legend.legendHandles)):
    legend.legendHandles[i]._sizes = [55]

plt.tight_layout()
plt.savefig("CorrEffect.png")




```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

Result = pd.read_csv("TreatmentEffect_Result.csv", index_col=False)
Result = Result.drop(Result.columns[0], axis=1)

sns.set_theme()
sns.set_style("ticks")
sns.set_context("paper")

Result = Result.set_index(['Treatment_Effect'])
print(Result.head())

markers = ['^', 'o', 's']
colors = sns.color_palette('tab10', n_colors=3)
ax,fig = plt.subplots(figsize = (6.5,4))
ax = sns.scatterplot(data=Result, markers=markers, palette=colors, edgecolor='black', alpha=0.8, s=55)
#ax.set(xlabel=r'$\rho$', ylabel='MSE')
ax.tick_params(axis='both', which='major', labelsize=13)
ax.set_xlabel(r'$\tau$', fontsize=14)
ax.set_ylabel('MSE', fontsize=14)
plt.xticks(fontsize=10.5)
plt.yticks(fontsize=10.5)
plt.grid()

handles, labels = ax.get_legend_handles_labels()
legend = ax.legend(handles, ['Oracle Access: $Y_{a_1}$', 'Observed Label: $Y_{a_3}$', 'Estimated Label: $\hat{Y}_{a_1}$'],edgecolor='black', loc='upper left', fontsize=13)
for i in range(len(legend.legendHandles)):
    legend.legendHandles[i]._sizes = [60]

plt.tight_layout()
plt.savefig("TreatmentEffect.png")
```
